```python
import os, glob

import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import tiffile as tif
```

```python
DIR = 'data/koeln'
```

```python
def parse_image(img_path: str) -> dict:   
    
    print(img_path)
    
    image = plt.imread(img_path).astype(np.float32)
    
    mask_path = img_path.replace('rgb.png', 'seg.tif')
    mask = tif.imread(mask_path).astype(np.int32)
    
    return {'image': image, 'segmentation_mask': mask}
```

```python
files = glob.glob(os.path.join(DIR,"*_rgb.png"))

SIZE = len(files)

geonrw = tf.data.Dataset.list_files(files)
geonrw = geonrw.shuffle(SIZE)
geonrw = geonrw.map(parse_image)
```

```python
import tensorflow_io as tfio

def parse_image_2(img_path: str) -> dict:
    image = tf.io.read_file(img_path)
    image = tf.image.decode_png(image, channels=3)
    image = tf.image.convert_image_dtype(image, tf.uint8)

    mask_path = tf.strings.regex_replace(img_path, "rgb.png", "seg.tif")
    mask = tf.io.read_file(mask_path)
    mask = tfio.experimental.image.decode_tiff(mask)

    return {'image': image, 'segmentation_mask': mask}
```

```python
files = glob.glob(os.path.join(DIR,"*_rgb.png"))

SIZE = len(files)

geonrw = tf.data.Dataset.list_files(files)
geonrw = geonrw.shuffle(SIZE)
geonrw = geonrw.map(parse_image_2)
```
